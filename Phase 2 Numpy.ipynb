# Numpy Basics

## Create Arrays

```python
import numpy as np

# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

# 2D Array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

## Array Properties

```python
print(arr1.shape)
print(arr2.shape)

print(arr1.dtype)
print(arr2.ndim)
```

## Indexing & Slicing

```python
print(arr1[0])
print(arr1[1:4])

print(arr2[0, 1])
print(arr2[:, 1])
```

## Mathematical Operations

```python
arr = np.array([10, 20, 30])

print(arr + 5)
print(arr * 2)
print(arr ** 2)
```

## Aggregation

```python
arr = np.array([10, 20, 30, 40])

print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
```

## Practice Task

```python
# Create an array of 10 numbers
# Find mean, max, min

arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

# Numpy Advanced

## Reshaping Arrays

```python
arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)
print(reshaped)
```

## Random Numbers

```python
print(np.random.rand(3))
print(np.random.randint(1, 100, 5))
```

## Boolean Filtering

```python
arr = np.array([10, 20, 30, 40, 50])

filtered = arr[arr > 25]
print(filtered)
```

## Practice Task

```python
# Create random array of 10 numbers between 1–50
# Print numbers greater than 25

arr = np.random.randint(1, 50, 10)

print(arr)
print(arr[arr > 25])
```
