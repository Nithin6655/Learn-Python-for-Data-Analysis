# Matplotlib Basics

## Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

## Multiple Lines

```python
y2 = [5, 15, 20, 25, 35]

plt.plot(x, y)
plt.plot(x, y2)

plt.title("Multiple Lines")
plt.show()
```

## Practice Task

```python
# Plot 2 lines:
# 1. Sales
# 2. Profit

days = [1, 2, 3, 4, 5]
sales = [100, 200, 150, 300, 250]
profit = [20, 50, 30, 70, 60]

plt.plot(days, sales)
plt.plot(days, profit)

plt.show()
```

## Bar & Histogram

### Bar Chart

```python
categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)
plt.title("Bar Chart")

plt.show()
```

## Histogram

```python
data = [10, 20, 20, 30, 30, 30, 40]

plt.hist(data)
plt.title("Histogram")

plt.show()
```

## Practice Task

```python
# Create histogram for marks

marks = [60, 70, 80, 90, 100, 70, 80, 85]

plt.hist(marks)

plt.show()
```

## Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.scatter(x, y)
plt.title("Scatter Plot")

plt.show()
```

## Seaborn Basics

```python
import seaborn as sns
import pandas as pd
```

## Load Dataset

```python
df = sns.load_dataset("tips")
print(df.head())
```

## Scatter Plot

```python
sns.scatterplot(x="total_bill", y="tip", data=df)

plt.show()
```

## Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=df)

plt.show()
```

## Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=df)

plt.show()
```

# Advanced Visualization

## Heatmap (Correlation)

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.show()
```

## Pairplot

```python
sns.pairplot(df)

plt.show()
```

# FINAL TASK

```python
# Load dataset
df = sns.load_dataset("tips")

# 1. Scatter: total_bill vs tip
sns.scatterplot(x="total_bill", y="tip", data=df)

# 2. Bar: day vs total_bill
sns.barplot(x="day", y="total_bill", data=df)

# 3. Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)

plt.show()
```
