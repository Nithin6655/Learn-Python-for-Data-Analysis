## Objective

Analyze Airbnb listings to identify pricing patterns, demand trends, and business insights.

## Step 1: Load & Understand Data

```python
import pandas as pd

df = pd.read_csv("airbnb.csv")

print(df.head())
print(df.info())
print(df.describe())
```

Questions:
What are the columns in the dataset?

Which columns are categorical vs numerical?

Are there missing values?

## Step 2: Data Cleaning

```python
# Missing values
print(df.isnull().sum())

# Fill or drop
df = df.dropna(subset=["price", "room_type"])

# Convert price if needed
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

Questions:

Why did you drop missing values instead of filling?

What happens if price has text values?

## Step 3: Basic Analysis

#### Average Price

```python
print(df["price"].mean())
```

Question:

What does average price tell you?

#### Most Common Room Type

```python
print(df["room_type"].value_counts())
```

Question:

Which room type is most preferred?

#### Price by Room Type

```python
print(df.groupby("room_type")["price"].mean())
```

Question:

Which room type is most expensive?

#### Top Locations

```python
print(df["neighbourhood_group"].value_counts())
```

Question:

Which area has most listings?

## Step 4: Visualization

### Price Distribution

```python
import matplotlib.pyplot as plt

plt.hist(df["price"], bins=50)
plt.title("Price Distribution")
plt.show()
```

Question:

Is the data skewed?

### Price by Room Type

```python
import seaborn as sns

sns.boxplot(x="room_type", y="price", data=df)
plt.show()
```

Question:

Why use boxplot here?

### Listings by Area

```python
sns.countplot(x="neighbourhood_group", data=df)
plt.show()
```

## Step 5: Advanced Analysis

### Expensive Listings

```python
expensive = df[df["price"] > 200]
print(expensive.head())
```

```python
Question:

What defines an “expensive” listing?
```

### Correlation

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```

Question:

Which factors affect price?

### Availability vs Price

```python
sns.scatterplot(x="availability_365", y="price", data=df)
plt.show()
```
