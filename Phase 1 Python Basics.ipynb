```python
# List
numbers = [1, 2, 3, 4, 5]
print(numbers)

# Tuple
coordinates = (10, 20)
print(coordinates)

# Set
unique_numbers = {1, 2, 2, 3}
print(unique_numbers)

# Dictionary
student = {"name": "John", "marks": 85}
print(student)
```

# Control Flow

```python
# If-Else
num = 10

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# For Loop
for i in range(5):
    print(i)
```

```python
# While Loop
count = 0

while count < 5:
    print(count)
    count += 1
```

```python
# Practice Task: Even numbers from list
numbers = [1, 2, 3, 4, 5, 6]

for num in numbers:
    if num % 2 == 0:
        print(num)
```

# Functions

```python
# Function to calculate average
def calculate_average(a, b, c):
    return (a + b + c) / 3

print(calculate_average(10, 20, 30))
```

```python
# Function to check prime number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
```

```python
# Practice Task
def greet(name):
    return "Hello " + name

print(greet("Nithin"))
```

# File Handling

```python
# Writing to a file
with open("sample.txt", "w") as file:
    file.write("Hello, this is a test file.")
```

```python
# Reading from a file
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Reading CSV file
import csv

with open("sample.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Nithin", 30])
    writer.writerow(["Rahul", 25])

with open("sample.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
