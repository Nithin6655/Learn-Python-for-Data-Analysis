# Pandas Series

## Create Series

```python
import pandas as pd

data = [10, 20, 30, 40]

s = pd.Series(data)
print(s)
```

## Custom Index

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
```

## Access Data

```python
print(s["a"])
print(s[0])   # position-based
```

## Operations on Series

```python
s = pd.Series([10, 20, 30, 40])

print(s + 5)
print(s * 2)
print(s.mean())
print(s.max())
```

## Filtering

```python
s = pd.Series([10, 20, 30, 40, 50])

print(s[s > 25])
```

## Series from Dictionary

```python
data = {
    "Nithin": 85,
    "Rahul": 90,
    "Anu": 88
}

s = pd.Series(data)
print(s)
```

## Handling Missing Values

```python
s = pd.Series([10, 20, None, 40])

print(s)

# Fill missing
s = s.fillna(s.mean())

print(s)
```

# Practice Task

```python
# Create a Series of 5 numbers
# Replace missing values with mean
# Print values greater than mean

s = pd.Series([10, 20, None, 40, 50])

s = s.fillna(s.mean())

print(s[s > s.mean()])
```

## Pandas DataFrame

## Create DataFrame

```python
import pandas as pd

data = {
    "Name": ["Nithin", "Rahul", "Anu"],
    "Age": [30, 25, 28],
    "Salary": [50000, 40000, 45000]
}

df = pd.DataFrame(data)
print(df)
```

## Basic Info

```python
print(df.head())
print(df.tail())

print(df.info())
print(df.describe())
```

## Column Selection

```python
print(df["Name"])

print(df[["Name", "Salary"]])
```

## Filtering Rows

```python
print(df[df["Age"] > 26])
```

## Add New Column

```python
df["Bonus"] = df["Salary"] * 0.1
print(df)
```

### Drop Column

```python
df = df.drop("Bonus", axis=1)
print(df)
```

### Handle Missing Values

```python
data = {
    "Name": ["Nithin", "Rahul", "Anu"],
    "Age": [30, None, 28]
}

df = pd.DataFrame(data)

print(df)

# Fill missing
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
```

### Group By

```python
data = {
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}

df = pd.DataFrame(data)

print(df.groupby("Department")["Salary"].mean())
```

```python
# Step 1: Create dataset
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [25, 30, 35, None],
    "Salary": [30000, 40000, 50000, 60000],
    "Department": ["IT", "HR", "IT", "HR"]
}

df = pd.DataFrame(data)

# Step 2: Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Step 3: Filter Salary > 40000
print(df[df["Salary"] > 40000])

# Step 4: Group by Department
print(df.groupby("Department")["Salary"].mean())
```
